Task 3. Work Experience Statistical Analysis

```python
work_exp = df['WorkExp']

work_exp_num = pd.to_numeric(work_exp, errors='coerce')

mean_exp = work_exp_num.mean()
median_exp = work_exp_num.median()
mode_exp = work_exp_num.mode()
print("Work experience statistics for respondents (WorkExp):")
print("Mean:", mean_exp)
print("Median:", median_exp)
print("Mode:", mode_exp.tolist())
```

Task 5. Python Popularity

Task 7. Geographic Analysis of Python Developer Compensation

```python
python_devs = df[df['LanguageHaveWorkedWith'].str.contains("Python", case=False, na=False)]

python_devs = python_devs[['Country', 'ConvertedCompYearly']].copy()

python_devs['ConvertedCompYearly'] = pd.to_numeric(python_devs['ConvertedCompYearly'], errors='coerce')

python_devs = python_devs.dropna(subset=['ConvertedCompYearly'])

country_comp = python_devs.groupby('Country')['ConvertedCompYearly'].agg(
    AverageCompensation='mean',
    MedianCompensation='median'
).reset_index()

country_comp = country_comp.sort_values(by='AverageCompensation', ascending=False)

print("Top 20 countries by average annual compensation for Python developers:")
country_comp.head(20)
```

Task 8. Education Analysis of Top Earners

```python
education_salary = df[['ConvertedCompYearly', 'EdLevel']].dropna(subset=['ConvertedCompYearly'])

education_salary_sorted = education_salary.sort_values(by='ConvertedCompYearly', ascending=False)

top5_education = education_salary_sorted.head(5)

top5_education[['EdLevel', 'ConvertedCompYearly']]
```

Task 9. Python Popularity by Age Group

```python
df_age = df.dropna(subset=['Age'])

df_age['UsesPython'] = df_age['LanguageHaveWorkedWith'].str.contains("Python", case=False, na=False)

python_by_age = df_age.groupby('Age')['UsesPython'].mean() * 100

python_by_age_df = python_by_age.reset_index()
python_by_age_df.rename(columns={'UsesPython': 'PercentPython'}, inplace=True)
python_by_age_df = python_by_age_df.sort_values(by='Age')

python_by_age_df
```

Task 10. Industry Analysis Among High-Paid Remote Workers

```python
df_comp = df[['ConvertedCompYearly', 'RemoteWork', 'Industry']].copy()

df_comp['ConvertedCompYearly'] = pd.to_numeric(df_comp['ConvertedCompYearly'], errors='coerce')

df_comp = df_comp.dropna(subset=['ConvertedCompYearly', 'RemoteWork', 'Industry'])

comp_75 = df_comp['ConvertedCompYearly'].quantile(0.75)

high_paid_remote = df_comp[
    (df_comp['ConvertedCompYearly'] > comp_75) &
    (df_comp['RemoteWork'].str.contains("Yes|Remote", case=False, na=False))
]

industry_counts = high_paid_remote['Industry'].value_counts().reset_index()
industry_counts.columns = ['Industry', 'Count']

print("Top 20 industries among high-paid remote workers:")
industry_counts.head(20)
```
